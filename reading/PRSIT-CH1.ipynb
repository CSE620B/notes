# Intro

_Ground based_ sensing uses in-situ measurements and surveying equipment. 

***Remote Senging***  is the topic of this course.  
It uses image data from a sensor

Common sensor types:
- Camera
- Radar
- LiDAR  (Lasers)

The sensors can be _aerial_ or _satelite_  (space) 

## Definitions of Remote Sensing

_Remote Sensing_ 
- Several definitions exist
- Common aspect: 
   - Data on Earth's surface acquired by a sensor that is not in contact with the object bing measured. 
   - Usually, this data is represented as an "image"  (I beleive the authors mean raster data)

Also _"Earth Observation (EO)_  
- spaceborn remote sensing

### Remote Sensing -- Integrating with Ground Data

_Data Fusion_ is an important part of RS

Link RS data with in-situe measurements (or other sources of GIS data)

###  RS --> "Map Like", Synoptic Data, Area Covering

From scattered in-situ measurements (e.g. bouys) plus spaceborn imagery, we can generate a wide areay holistic view. 

The data can be _geolocated_ or _georeferenced_ in order to link measurements to map coordinates

Georeferenced data can be rendered as a map, 
1.  Pseudocolor rendering maps measurements to colors
2.  Hillsheading, Elevation Maps, Countour Maps
3.  Data may need to be interpolated or summarized (heat maps)
